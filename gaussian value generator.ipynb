```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import math
import random
pi=math.pi
def prob(b,x,y):
    a=((b-x)/y)**2
    f=1/(y*(2*pi))
    f*=math.exp(-1/2*(a))
    return f
```

```python
def get_speed(num,speed,variance):
    f=[]
    for i in range(num):
        arr=[i for i in range(-variance,variance+1)]
        n=int(random.random()*len(arr))
        num_list=[speed+arr[n]*random.random() for i in range(50)]
        value=num_list[0]
        p_max=0
        for i in range(len(num_list)):
            p_value=prob(num_list[i],speed,variance)
            if p_value>p_max:
                p_max=p_value
                value=num_list[i]
        f.append(value)
    return f
```

```python
a=get_speed(1000,60,50)
```

```python
a=np.asarray(a)
```

```python
a.mean()
```

```python
a.var()
```
